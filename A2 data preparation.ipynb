```python
def Output_Topic(file):
    file_object = open(file, "r",encoding='utf-8')
    lines = file_object.readlines()

    # list for storing topic ID
    topic = []
    for top in lines:
        if "ID tr" in top:
            topic.append(top)
            
    file_object.close()
    return(topic)
topic = Output_Topic(txt_file_path)
# topic
```
